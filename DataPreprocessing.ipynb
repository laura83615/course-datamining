## Adding Full/Empty State Column
Percent_Full_State column is derived from Percent_Full

Available_State column is derived from Available_Docks and Available_Bikes column

```python
def percent_full_state(row):
    if row['Percent_Full'] < 20:
        return 0
    elif row['Percent_Full'] < 80:
        return 1
    else:
        return 2

station_pd['Percent_Full_State'] = station_pd.apply (lambda row: percent_full_state (row),axis=1)
station_pd.head()
```

# Process all the files in "split_2015/"

```python

from os import listdir
from os.path import isfile, join

# get list of file in directory "split_2015/"
mypath = "split_2015/"
onlyfiles = [f for f in listdir(mypath) if isfile(join(mypath, f))]

def percent_full_state(row):
    if row['Percent_Full'] < 20:
        return 0
    elif row['Percent_Full'] < 80:
        return 1
    else:
        return 2

def available_state(row):
    if row['Available_Bikes'] <= 3:
        #few bikes are available therefore it is nearly empty state
        return 0
    elif row['Available_Docks'] <= 3:
        #few docks are available therefore it is nearly full state
        return 2
    else:
        return 1

FMT = '%m/%d/%Y %I:%M:%S %p'

storepath = 'processed_2015/'

colnames = []
for f in onlyfiles:
    if len(colnames) == 0:
        station_pd = pd.read_csv(mypath+f)
        colnames = station_pd.columns.values.tolist()
    else:
        station_pd = pd.read_csv(mypath+f, names = colnames)
    
    #adding weekday 
    station_pd['Timestamp'] = pd.to_datetime(station_pd['Timestamp'])
    station_pd['Weekday'] = station_pd['Timestamp'].dt.dayofweek
    
    #adding Percent_Full_State
    station_pd['Percent_Full_State'] = station_pd.apply (lambda row: percent_full_state (row),axis=1)
    
    #adding Available_State
    station_pd['Available_State'] = station_pd.apply (lambda row: percent_full_state (row),axis=1)
    
    station_pd.to_csv(storepath+f)
```
