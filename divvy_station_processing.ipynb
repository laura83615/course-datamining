```python
with open("hao/2015/Divvy_Bicycle_Stations_-_Historical.csv",'r') as f:
    i = 1
    stationName_id_mapping = dict();
    station_id_count = 0;
    file_count = 0;
    file_path = "hao/2015/split_2015/split_2015_" + str(file_count)+ ".csv"
    for line in f:
        #print(line)
        words = line.split(",")
        #print(words)
        
        if stationName_id_mapping.get(words[2]) == None:
            stationName_id_mapping[words[2]] = station_id_count
        station_id_count += 1
        words[2] = str(stationName_id_mapping[words[2]])
        words = words[:3] + words[4:10] + words[13:14]
        list_to_string = ",".join(words)
       #print(list_to_string)
        with open(file_path,'a') as fileout:
            fileout.write(list_to_string)
        
        i+=1
        if i >= 100000:
            file_count += 1
            i = 1
            file_path = "hao/2015/split_2015/split_2015_" + str(file_count)+ ".csv"
    
```

```python
# merge data frames
dfs = []
for i in range(0, 22):
    file_path = "hao/processed_2015/split_2015_" + str(i)+".csv"
    tfd = pd.read_csv(file_path, dtype = {'ID': int,'Timestamp': str,'Station_ID': int,'Total_Docks': int,'Docks_in_Service': int,'Available_Docks': int,
                                                'Available_Bikes': int,'Percent_Full': int,'Status': str,'Record': int, 'Weekday': int,'Percent_Full_State': int,'Available_State': int})
    dfs.append(tfd)
    #print(tfd.head)
    #break
divvy_station = pd.concat(dfs, ignore_index=True)
#print(divvy_station.head)
print(divvy_station.shape)
#print(divvy_station.head[3])
```
